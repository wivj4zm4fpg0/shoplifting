```python
import random
```

```python
hoge = {}
rand = list(range(50))
random.shuffle(rand)
index = 0
```

```python
for i in rand:
    hoge[i] = index
    index += 1

print(hoge)
```

```python
import operator
```

```python
sorted_hoge = sorted(hoge.items(), key=operator.itemgetter(0))
```

```python
print(sorted_hoge)
```
